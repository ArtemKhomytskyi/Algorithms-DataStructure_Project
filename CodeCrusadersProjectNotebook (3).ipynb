### Extraction of recency, frequency and monetary values

```python
current_date = (1, 5, 2023)
days_since_last_purchase={}
# Calculate and save the number of days since each student's latest purchase
for purchase in dataset:
    purchase_date = purchase['Date']
    days_since_purchase = (current_date[0] - purchase_date[0]) + (30 * (current_date[1] - purchase_date[1])) + (365 * (current_date[2] - purchase_date[2]))
    name = purchase["Name"]
    if name not in days_since_last_purchase:
#Check if the student name exists as a key in the purchases_by_student dictionary.
        days_since_last_purchase[name] = days_since_purchase
    elif days_since_purchase<days_since_last_purchase[name]:
        days_since_last_purchase[name]=days_since_purchase
for student, days in days_since_last_purchase.items():
    print(f"{student}: {days} since last purchase")
# Print the updated dataset
for student in students:
  student['Recency'] = days_since_last_purchase[student['Name']]
```

```python
purchases_by_student = {}
for purchase in dataset:
    name = purchase['Name']
    if name not in purchases_by_student:
#Check if the student name exists as a key in the purchases_by_student dictionary.
        purchases_by_student[name] = 0
#If the student name does not exist, add it to the purchases_by_student dictionary and initialize the purchase count to 0.
    purchases_by_student[name] += 1
#Increment the purchase count for the corresponding student name in the purchases_by_student dictionary.

for student, purchase_count in purchases_by_student.items():
    print(f"{student}: {purchase_count} purchases")
# Print the total number of purchases made by each student
for student in students:
  student['Frequency'] = purchases_by_student[student['Name']]
```

```python
amount_spent_by_student = {}
for purchase in dataset:
    student_name = purchase['Name']
    item_price = purchase['Unit Price']
    item_quantity = purchase['Quantity']
    total_amount = item_price * item_quantity
    
    if student_name not in amount_spent_by_student:
        amount_spent_by_student[student_name] = 0.0
#If the student name does not exist, add it to the amount_spent_by_student dictionary and initialize the total amount to 0.0.
    amount_spent_by_student[student_name] += total_amount
#Add the total amount spent for the current purchase to the corresponding student's total amount in the amount_spent_by_student dictionary.

for student, total_amount in amount_spent_by_student.items():
    print(f"{student}: Total amount spent = ${total_amount:.2f}")
#Print the student name and the corresponding total amount spent by that student formatted as a currency with two decimal places.

for student in students:
  student['Monetary'] = amount_spent_by_student[student['Name']]
```
